```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}
url = "https://www.amazon.in/s?k=home+decor+items&crid=1MC34Q5L99KBN&qid=1709367592&sprefix=home+decor+%2Caps%2C283&ref=sr_pg_1"

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "lxml")


name=soup.find_all("div",class_="puis-card-container s-card-container s-overflow-hidden aok-relative puis-expand-height puis-include-content-margin puis puis-v10dnrav6sitdx2esf8fqwtjtbs s-latency-cf-section puis-card-border")

image=[]
product_name=[]
product_rating=[]
total_rating=[]
last_month_purchaging=[]
product_price=[]
for i in name:
  image_tag = i.find("img", class_="s-image")
  image.append(image_tag['src'] if image_tag else np.nan)
  product_name.append(i.find("span",class_="a-size-base-plus a-color-base a-text-normal").text.strip())
  product_rating.append(i.find("span",class_="a-icon-alt").text.strip())
  total_rating.append(i.find("span",class_="a-size-base s-underline-text").text.strip())
  last_month_purchaging.append(i.find("span",class_="a-size-base a-color-secondary").text.strip())
  product_price.append(i.find("span",class_="a-price-whole").text.strip())

dict={
    "image":image,
"product_name":product_name,
"product_rating":product_rating,
"total_rating":total_rating,
"last_month_purchaging":last_month_purchaging,
"product_price":product_price
}

df1=pd.DataFrame(dict)
df1
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}
url = "https://www.amazon.in/s?k=home+decor+items&page=2&crid=1MC34Q5L99KBN&qid=1709374010&sprefix=home+decor+%2Caps%2C283&ref=sr_pg_2"

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "lxml")

name=soup.find_all("div",class_="puis-card-container s-card-container s-overflow-hidden aok-relative puis-expand-height puis-include-content-margin puis puis-v10dnrav6sitdx2esf8fqwtjtbs s-latency-cf-section puis-card-border")

image=[]
product_name=[]
product_rating=[]
total_rating=[]
last_month_purchaging=[]
product_price=[]
for i in name:
  image_tag = i.find("img", class_="s-image")
  image.append(image_tag['src'] if image_tag else np.nan)
  product_name.append(i.find("span",class_="a-size-base-plus a-color-base a-text-normal").text.strip())
  product_rating.append(i.find("span",class_="a-icon-alt").text.strip())
  total_rating.append(i.find("span",class_="a-size-base s-underline-text").text.strip())
  last_month_purchaging.append(i.find("span",class_="a-size-base a-color-secondary").text.strip())
  product_price.append(i.find("span",class_="a-price-whole").text.strip())

dict={
    "image":image,
"product_name":product_name,
"product_rating":product_rating,
"total_rating":total_rating,
"last_month_purchaging":last_month_purchaging,
"product_price":product_price
}

df2=pd.DataFrame(dict)
df2
```

```python
df=pd.concat([df1,df2])
```
